```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv('titanic_x_y_train.csv')
df.head()
```

```python
df.drop('Name',axis = 1,inplace = True)
df.drop('Cabin',axis = 1,inplace = True)
df.drop('Ticket',axis = 1,inplace = True)
df.Age.fillna(df.Age.mean(),inplace = True)
```

```python
df.head()
```

```python
def f(a):
    if a == 'male':
        return 0
    else:
        return 1
df['Sex'] = df.Sex.apply(f)
df.head()
```

```python
df.drop('Embarked',axis = 1,inplace = True)
df.head()
```

```python
data = df.values
data
data.shape
```

```python
def cost(x,y,m):
    M = len(x)
    total_cost = 0
    for i in range(M):
        X = x[i,:]
        Y = y[i]
        for j in range(7):
            total_cost += (1/M)*(-(Y)*math.log((1/(1 + math.exp(-(m*X).sum())))) - (1 - Y)*math.log(1 - (1/(1 + math.exp(-(m*X).sum())))))
        return total_cost
    
def descent(x,y,learning_rate,m):
    m_slope = np.zeros(len(m))
    M,N = x.shape
    for i in range(M):
        X = x[i,:]
        Y = y[i]
        for j in range(7):
            m_slope[j] += (-1/M)*(Y - (1/(1 + math.exp(-(m*X).sum()))))*X[j]
        new_m = m - learning_rate*m_slope
    return new_m
            
def gd(x,y,learning_rate,number_of_iteration):
    m = np.zeros(7)
    for i in range(number_of_iteration):
        m = descent(x,y,learning_rate,m)
        print(i, 'cost : ', cost(x,y,m))
    return m

def run():
    M,N = data.shape
    x = data[:,:-1]
    y = data[:,-1]
    x = np.append(x, np.ones([M,1]),axis = 1)
    number_of_iteration = 1000
    learning_rate = 0.001
    m = gd(x,y,learning_rate,number_of_iteration)
    print(m)
```

```python
run()
```
