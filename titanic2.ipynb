```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('titanic_x_y_train.csv')
df2 = pd.read_csv('titanic_x_test.csv')
df.head()
df2.head()
```

```python
df.drop('Name',axis = 1,inplace = True)
df.drop('Cabin',axis = 1,inplace = True)
df.drop('Ticket',axis = 1,inplace = True)
df.Age.fillna(df.Age.mean(),inplace = True)
df2.drop('Name',axis = 1,inplace = True)
df2.drop('Cabin',axis = 1,inplace = True)
df2.drop('Ticket',axis = 1,inplace = True)
df2.Age.fillna(df.Age.mean(),inplace = True)
```

```python
df.head()
df2.head()
```

```python
def f(a):
    if a == 'male':
        return 0
    else:
        return 1
df['Sex'] = df.Sex.apply(f)
df2['Sex'] = df2.Sex.apply(f)
df2.head()
```

```python
df.drop('Embarked',axis = 1,inplace = True)
df2.drop('Embarked',axis = 1,inplace = True)
df2.head()
```

```python
data = df.values
data2 = df2.values
x = data[:,:-1]
y = data[:,-1]
```

```python
clf = LogisticRegression()
clf.fit(x,y)
```

```python
y = clf.predict(data2)
```

```python
np.savetxt("predict.csv",y,delimiter = ',')
```
